# Phase 6 — interactive dashboard (no automation)

Regenerates a single self-contained HTML dashboard at
`outputs/dashboard/index.html` from the processed parquets. The charts are
**live SVG** (hand-rolled vanilla JS, no CDN, no server): hover any chart for
the exact NDVI and date, toggle years in the legend, hover a zone pixel for its
season-mean NDVI, sort the features table, and switch blocks with the tabs.
Grower KPIs (latest reading, this season's peak, green-up, low-vigor alerts)
sit up top. The data is embedded as JSON, so the file works offline from
`file://`.

This is **the dashboard only** — the weekly scheduled run (cron / Task
Scheduler) from the Phase 6 spec is intentionally not built. To refresh:
re-run notebooks 02–05 to update the parquets, then run this, or from a shell
`python -m vigor.webdash`.

Inputs (from `data/processed/`): `block_timeseries.parquet` (required),
`season_features.parquet`, `vigor_zones.parquet`, `vigor_alerts.parquet`.
Missing optional inputs simply drop their section.

> The earlier static-image builder is still available as `vigor.dashboard`
> (`python -m vigor.dashboard`) if a print-friendly, no-JavaScript version is
> ever needed.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from vigor import webdash

out = webdash.build_dashboard(
    processed_dir=PROJECT_ROOT / "data" / "processed",
    blocks_path=PROJECT_ROOT / "data" / "blocks.geojson",
    out_dir=PROJECT_ROOT / "outputs" / "dashboard",
)
print(f"interactive dashboard -> {out}")
print(f"size: {out.stat().st_size / 1024:.0f} KB")
```
